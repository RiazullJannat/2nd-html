```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./sample_data/titanic_data_updated.csv")
print(f"Samples: {df.shape[0]}")
print(f"Features: {df.shape[1]}")
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

#Uni-varitente Analysis (categorical)

```python
sns.countplot(data = df, x=df["Survived"])
```

```python
survived_count = (df["Survived"].value_counts() / df.shape[0] *100)
print(survived_count)
```

```python
plt.pie(survived_count, labels=survived_count.index, autopct='%1.1f%%')
plt.show()
```
